### Fetch Watch list

```python

url = PROD_API_URL_BASE + '/omni/internal/dashboard/get/watchlist/all?limit=20&avoidCounting=false&offset='

min_date = 1784613600000

#1784613600000 July 21st
#1784765379389000
payload = {
'_createdAt': {'$gte': 1784613600000}
}



more_pages = True
limit = 20
current_page = 0
total_records = 0

my_file = open("my_records.csv", "w")
my_file.write('id,created_at,n\n')
temp_data = []


while ( more_pages == True ):

    headers = {
        "accept": "application/json",
        "api-version": "1.0",
        "content-type": "application/json",
        'x-incode-hardware-id': HARDWARE_ID,
        "x-api-key": API_KEY,
        'origin': PROD_API_URL_BASE
    }


    #print(url + str(limit * current_page))
    
    response = requests.post( ( url + str(limit * current_page) ), json=payload, headers=headers)
    current_page+=1

    raw_data = response.json()
    if( current_page == 1):
        total_records = raw_data.get('total')
        more_pages = raw_data.get('more')
    else:
        if( current_page * limit > total_records ):
            more_pages = False
    
    print( more_pages)

  
    print('%s of %s'%((limit * current_page), total_records))

    records = raw_data.get('incodeWatchlistRecords')

    for line in records:
        if line['_createdAt'] < min_date:
            print("warning, date below the minimal")
            print('% s,% s' %( line['_id'] , line['_createdAt'] ))
        my_file.write('% s,% s,"% s"\n' %( line['_id'] , line['_createdAt'], line['name'] ))
        #pprint.pp(line)
        temp_data.append([line['_id'] , line['_createdAt'], line['name']])

#pprint.pp(temp_data)
print(len(temp_data))
my_file.close()

#pprint.pp(response.json())





```
