### 1. Begin Verification

#### 1.1 Start onboarding `/omni/start` 

Creates a new verification session. Returns the session token and interviewId used in all subsequent calls. 

````python
url = "https://demo-api.incodesmile.com/omni/start"

headers = {
    "accept": "application/json",
    "api-version": "1.0",
    "content-type": "application/json",
    "x-api-key": API_KEY
}

payload = {"configurationId": CONFIGURATION_ID, "externalCustomerId": CUSTOMER_ID,
           "customFields": {"firstName": "John", "lastName": "Smith"}}

response_start = requests.post(url, json=payload, headers=headers)

interview_id = response_start.json()['interviewId']
hardware_id = response_start.json()['token']

display(Markdown(f"\
| Field | Value            |\n\
|:------|:----------------|\n\
| interviewId | `{interview_id}` |\n\
| token | `{hardware_id}` |"))
# pprint.pp(response.json())
````

#### 1.2 Get On boarding URL `/omni/oboarding-url`

Returns the hosted onboarding URL for the current session. 

````python
url = "https://demo-api.incodesmile.com/omni/onboarding-url"

headers = {
    "accept": "application/json",
    "api-version": "1.0",
    "x-api-key": API_KEY,
    "X-Incode-Hardware-Id": hardware_id
}

response_url = requests.get(url, headers=headers)

onboarding_url = response_url.json()['url']

pprint.pp(onboarding_url)
#pprint.pp(response.json())

#display(Markdown(f"[Onboarding URL]({onboarding_url})"))

IFrame(src=onboarding_url, width="100%", height=300)

````

#### 1.3 Get flow information `/omni/onboarding/flow`

Returns the full configuration of a Flow, including all enabled modules and their settings. 

````python
url = "https://demo-api.incodesmile.com/omni/onboarding/flow"

headers = {
    "accept": "application/json",
    "api-version": "1.0",
    "x-api-key": API_KEY,
    "X-Incode-Hardware-Id": hardware_id,


}

payload = {'id': CONFIGURATION_ID}

response_flow = requests.get(url, headers=headers, params=payload)

display(Markdown(f"```json\n\
{pprint.pformat( response_flow.json() )} \n\
```"))

# pprint.pp( response.json() )
````

### 1.4  Merge Responses 

After calling sections 1.1, 1.2, and 1.3, merge the three response objects into a single payload. Then add the following properties from your key store before returning the response to the client. Then add `apiUrl` and `consentId`to the payload.

````python
merged_responses = response_start.json() | response_url.json() | response_flow.json()

merged_responses['apiUrl'] = '';
merged_responses['consentId'] = merged_responses['flowModules'][0]['configuration']['combinedConsents'];


display(Markdown(f"```json\n\
{pprint.pformat( merged_responses )} \n\
```"))
````
